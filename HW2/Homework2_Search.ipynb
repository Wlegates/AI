```python
#Problem 1 find friends
from collections import deque


#The network represented by an adjacency list
Network = {
    "bob": ["Rob", "Richard", "Pam"],
    "Pam": ["Rodger", "Peter"],
    "Rodger": ["Anna"],
    "Peter" : ["Amy"]
}


def Findfriends(Graph, user, K):
  visited = set()
  queue = deque([(user,0)])
  visited.add(user)

  while queue:
    vertex, depth = queue.popleft()

    #only prints the people at the depth of k
    if depth == K:
      print(vertex, end=" ")

#if the  depth is not found then the loop continues
    if depth < K:
      for neighbour in Graph.get(vertex, []):
        if neighbour not in visited:
          visited.add(neighbour)
          queue.append((neighbour, depth + 1))

#Main function
def Main():
  Findfriends(Network, "bob", 3)

if __name__ == "__main__":
  Main()
```

```python
#Problem 2 Missing connections
from collections import deque

Network = {
    "Sophia": ["Maya", "Maria"],
    "Maria": ["Maya", "Sophia", "Adam", "David"],
    "Maya": ["Maria", "Sophia", "David"],
    "David": ["Maya", "Maria"],
    "Adam": ["Maria"]
}

def Potentialfriends(graph, user):
  visited = set()
  queue = deque([(user, 0)])
  visited.add(user)
  Newfriends = []

  while queue:
    vertex, depth = queue.popleft() # Unpack both vertex and depth

    if depth == 2:
      Newfriends.append(vertex)

    if depth < 2:
      for neighbour in graph.get(vertex, []):
        if neighbour not in visited:
          visited.add(neighbour)
          queue.append((neighbour, depth + 1))

  return Newfriends # Return the list of potential friends

def Main():
  print(Potentialfriends(Network, "Sophia"))

  print(Potentialfriends(Network, "Adam"))

  print(Potentialfriends(Network, "David"))

if __name__ == "__main__":
  Main()
```
